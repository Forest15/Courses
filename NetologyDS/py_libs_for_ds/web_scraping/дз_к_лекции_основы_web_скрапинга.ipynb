```python
from bs4 import BeautifulSoup
import pandas as pd
import requests as re
```

```python
def get_all_links(params):
    url = 'https://habr.com/ru/search/'
    
    params = {
        'q': params
    }
    
    rows = {'date': [], 'title': [], 'href': []}
    result_df = pd.DataFrame()
    
    for param in params:
        res = re.get(url, params)
        soup = BeautifulSoup(res.text)
        art_dt_rs = soup.find_all('a', class_ = 'tm-article-datetime-published tm-article-datetime-published_link')
        art_title_href_rs = soup.find_all('h2', class_ = 'tm-title tm-title_h2')
    
        for art_dt_item, art_title_href_item in zip(art_dt_rs, art_title_href_rs):
            res_row = {
                'date': art_dt_item.text, 
                'title': art_title_href_item.find("a").find("span").text, 
                'href': f"habr.com{art_title_href_item.find('a').get('href')}"}
            row_df = pd.DataFrame(res_row, index=[0])
            result_df = pd.concat([result_df, row_df], ignore_index=True)
        
    
    result_df.drop_duplicates(inplace = True)
    return result_df
```

При составлении строки-аргумента query из нескольких слов, в местах пробелов нужно ставить знак плюс.

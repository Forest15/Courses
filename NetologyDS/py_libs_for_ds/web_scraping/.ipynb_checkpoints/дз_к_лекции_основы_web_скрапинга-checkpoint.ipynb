```python
from bs4 import BeautifulSoup
import pandas as pd
import requests as re
```

```python
def get_all_links(params):
    url = 'https://habr.com/ru/search/'
    
    params = {
        'q': params
    }
    
    res = re.get(url, params)
    soup = BeautifulSoup(res.text)
    art_dt_rs = soup.find_all('a', class_ = 'tm-article-datetime-published tm-article-datetime-published_link')
    art_title_href_rs = soup.find_all('h2', class_ = 'tm-title tm-title_h2')
    
    rows = {'date': [], 'title': [], 'href': []}
    
    for art_dt_item in art_dt_rs:
        rows['date'].append(art_dt_item.text)
            
    for art_title_href_item in art_title_href_rs:
        rows['title'].append(art_title_href_item.find("a").find("span").text)
        rows['href'].append(f"habr.com{art_title_href_item.find('a').get('href')}")
    
    result_df = pd.DataFrame(rows)
    result_df.drop_duplicates
    return result_df
```

При составлении строки-аргумента query из нескольких слов, в местах пробелов нужно ставить знак плюс.

```python
pd.set_option('display.max_colwidth', None) # Чтобы в df текст названия отображался полностью

get_all_links("анализ+данных")
```

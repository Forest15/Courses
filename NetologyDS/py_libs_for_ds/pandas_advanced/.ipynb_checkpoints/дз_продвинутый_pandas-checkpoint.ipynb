## Задание 1

```python
import pandas as pd
```

```python
log_df = pd.read_csv('visit_log.csv', sep=';')
log_df.head()
```

```python
log_df.info()
```

```python
log_df.traffic_source.unique()
```

```python
def ts_func(log_df):
    if 'yandex' in log_df["traffic_source"] or 'google' in log_df["traffic_source"]:
        return 'organic'
    elif ('paid' in log_df["traffic_source"] or 'email' in log_df["traffic_source"]) and ('russia' in log_df["region"].lower()):
        return 'ad'
    elif ('paid' in log_df["traffic_source"] or 'email' in log_df["traffic_source"]) and ~('russia' in log_df["region"].lower()):
        return 'other'
    else:
        return log_df.traffic_source
```

```python
log_df["source_type"] = log_df.apply(ts_func, axis = 1)
```

```python
log_df.source_type.value_counts()
```

## Задание 3

```python
rating_df = pd.read_csv('./ml-latest-small/ratings.csv')
rating_df.head()
```

```python
rating_df.info()
```

```python
rating_grp = rating_df.groupby("userId").agg({'movieId': 'count', 'timestamp': ['min', 'max']}).loc[rating_grp["movieId"]["count"] > 100]
```

```python
rating_grp
```

```python
rating_grp["life_time"] = (rating_grp["timestamp"]["max"] - rating_grp["timestamp"]["min"]) /24 /3600
```

```python
rating_grp.life_time.mean()
```

### Задание 4

Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

### Подзадание 1

Последняя таблица включает адреса всех клиентов. Она будет главной (левой) при объединении.

```python
joined_table_rzd = client_base.merge(rzd, how = 'left', on = 'client_id')[["client_id", "rzd_revenue"]]
```

```python
joined_table_rzd
```

```python
joined_table_auto = client_base.merge(auto, how = 'left', on = 'client_id')[["client_id", "auto_revenue"]]
```

```python
joined_table_auto
```

```python
joined_table_air = client_base.merge(air, how = 'left', on = 'client_id')[["client_id", "air_revenue"]]
```

```python
joined_table_air
```

```python
joined_table_cost = joined_table_auto.merge(joined_table_air, how = 'left', on = 'client_id').merge(joined_table_rzd, how = 'left', on = 'client_id').fillna(0)
```

```python
joined_table_cost
```

### Подзадание 2

```python
joined_table_cost_with_addr = joined_table_cost.merge(client_base, how = 'inner', on = 'client_id')
```
